Matplotlib — библиотека на языке программирования Python для визуализации данных двумерной (2D) графикой (3D графика также поддерживается). Получаемые изображения могут быть использованы в качестве иллюстраций в публикациях

```python
import matplotlib 
import matplotlib.pyplot as plt


plt.plot([1,14,7,5,4])
# показываем окно графика в блакнотах к этой функции нет необходимости
plt.show()
```

pyplot - основной инструмент работы с графиками  в mathplotlib
matplotlib создан для работы с numpy наилучшим образом именно со структурами этого модуля 

```python
import numpy as np
x = np.array([1,1,3,4,5])
y = np.array([1,4,6,8,12])
plt.plot(x,y)
```

```python
y = np.array ([0,5,8])
x = np.array([a*a for a in y])
# возможно создать сетку, она рисуется только при вызове метода
plt.grid()
plt.plot(x,y)
```

```python
x1 = np.array([1,2,3,4,5])
y1 =np.array([i*i for i in x1])
x2 = np.array([1,2,3,4,5,6,7])
y2 =np.array([i*i*i for i in x2])
#Возможно вывести несколько графиков просто передав множеста 
plt.plot(x1,y1,x2,y2)
```

```python
x1 = np.array([1,2,3,4,5])
y1 = np.array([i*i for i in x1])
x2 = np.array([1,2,3,4,5,6])
y2 = np.array([i**3 for i in x2])
x3 = np.array([1,2,3,4])
y3 = np.array([i**4 for i in x3])

# Можно указать тип линии и цвет
plt.plot(x1,y1,'o--r',x2,y2,'x:g')
plt.plot(x3,y3,'*-.b')
plt.grid()
```

```python
x1 = np.array([1,2,3,4,5])
y1 = np.array([i*i for i in x1])

line = plt.plot(x1,y1) 
#можно задать отельно  параметры графика используя метод setp
plt.setp(line,linestyle = '-.',linewidth = 2, color =(0.1,0.5,0.8,0.4), marker ='v' )
help(plt.plot)
```

```python
x = np.arange(-2*np.pi,2*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y)
# можно закрасить график  методом fill_between
plt.fill_between(x,y, where= y<0, color ='r')
plt.fill_between(x,y, where= y>0, color ='g')
```
