NumPy — библиотека с открытым исходным кодом для языка программирования Python. Возможности:

поддержка многомерных массивов (включая матрицы);
поддержка высокоуровневых математических функций, предназначенных для работы с многомерными массивами.

```python
import numpy as np 
# help(np.array)
a = np.array([1,2,4])
print(a)

#Все элементы массива должны иметь одинаковый тип
a = np.array([1,2,'4',True])
print(a)
# тип данных будет преобразовываться автоматически
a[2] = 6.43
print(a)
# можно заполнить массив элементами другого массива копировав их
a = a[[1,1,3,1,1,1,1,3,1]]
print(a)
# возможно создавать массивы на основе логических операций 
a = a[[True,True,False,True,True,True,True,True,True]]
print(a)
""" Метод reshape позволяет создать новый массив с произвольным количеством строк
и столбцов"""
b = a.reshape(4,2)
print(b)
```

```python
# Способы формирования массивов 

#просто перечислить элементы 
a = np.array([1,2,3,4])
a = np.array([1,2,3,4],'float')
print(a)
a = np.array([1,2,3,400],dtype='uint32')
print(a)
a = np.int8(a)
print(a)
```

```python
# первый элемент может быть либо списком либо кортежем 
a = np.array('hello')
a = np.array([[2,4],[4,6],[6,8]])
print(a)
b = np.array([[a*2],[a*3],[a*4]])
print('b=', b)
print('b=', b[0])
```

```python
# Способы создания массивов
a = np.array([1]*10)
print(a)
# возможно использовать механизм генерации списков 
a = np.array([i*2 for i in range(12)])
print(a)
# Создать массив заполненный случайно 
a = np.empty(34, 'int16')
print(a)
# Создает матрице m  на n  c  единицами на "диагонали"
a = np.eye(4,4,-1,'int')
print(a)
# создает диагональную матрицы размерности n
a = np.identity(12)
print(a)
# заполнить нулями или единицами или конкретными значениями
a = np.ones(125)
a = np.zeros(62)
a = np.full(62, 666)
print(a)
```

```python
# создает матрицу
a = np.mat([(1,2,3), (4,5,6)])
a = np.mat('1 2 3 4 5')
a = np.mat('1,2,3,4,5')
a = np.mat('1,2; 4,5')
print(a)
print('___________')
# создает диагональную матрицу
a = np.diag([1,2,3,4])
print(a)
# или возвращает главную даигональ
a = np.diag([(8,2,3), (4,8,6), (7,8,8)])
print(a)
# можно создать даигональную матрицу из списка 
a = np.diagflat([(8,2,3), (4,8,6), (7,8,8)])
print(a)
# треуголную матрицу можно создать так
a = np.tri(4,4)
print(a)
# или привести матрицу к диагональному виду 
a = np.tril(a.transpose())
print(a)
# можно сделать обратную диагональную матрицу 
a = np.mat([[3]*3,[3]*3,[3]*3])
a = np.triu(a)
print(a)
```

```python
# формирование числовых диапазонов 
a = np.arange(4)
print(a)
a = np.arange(0,np.pi,0.1)
print(a)
#Линейная прогрессия 
a= np.linspace(0,np.pi*2,100)
print(a)
#логарифмическая
a= np.logspace(7,18,12)
print(a)
#геометрическая
a= np.geomspace(7,18,12)
print(a)
# Можно создать копию массива
b = np.copy(a)
```

```python
a = np.fromiter('hello', dtype ='<U1')
print(a)
a= np.fromstring('1 2 3 ', dtype = 'int8', sep=' ')
print(a)
```
