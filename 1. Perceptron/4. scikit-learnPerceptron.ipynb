```python
from sklearn import datasets
import numpy as np

iris = datasets.load_iris()
X = iris.data[:,[2,3]]
y = iris.target
print('Метки классов', np.unique(y))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, 
                                                 random_state =1, stratify=y)
print("количество меток в y",np.bincount(y))
print("количество меток в y_train",np.bincount(y_train))
print("количество меток в y_test",np.bincount(y_test))
```

```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

plt.subplots(1,1)
plt.scatter(x=X_test[:,0] ,y=X_test[:,1],marker="o")
plt.subplots(1,1)
plt.scatter(x=X_test_std[:,0] ,y=X_test_std[:,1],marker='x')
plt.show()
```

```python
from sklearn.linear_model import Perceptron
ppn = Perceptron(n_iter_no_change=40 ,eta0=0.01, random_state=1)
ppn.fit(X_train_std,y_train)
y_pred = ppn.predict(X_test_std)
print('Неверно классифицированные образцы %d' % (y_test != y_pred).sum() )
print(np.mean(y_pred==y_test)-1)
```
